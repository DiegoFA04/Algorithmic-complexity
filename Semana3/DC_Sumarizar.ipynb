# Suma de Valores de una Lista (Arreglo)

```python
import random

def sumarize(a, i, f):
  if i == f:
    return a[i]
  else:
    mid = (i + f) // 2
    s1 = sumarize(a, i, mid)
    s2 = sumarize(a, mid + 1, f)
    return s1 + s2

def main():
  arr = list(range(30)) #Genera elementos ordenados [0,1,2,...,29] 
  #print(arr)          
  random.shuffle(arr)   #Barajar la secuencia
  print("Arreglo: ", arr)
  
  print("Suma: ", sumarize(arr, 0, len(arr) - 1))

main()
```
